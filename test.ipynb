```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import pickle
from Optional_EDA import *
from EDA import *

class TestProgram:
    def __init__(self):
        self.df = None

    def read_data(self, file_path):
        self.df = pd.read_csv(file_path)
        print("Data loaded successfully.")

    def display_menu(self):
        while True:
            print("\n--- Main Menu ---")
            print("1. Explore Data")
            print("2. Descriptive Statistics")
            print("3. Frequency distribution Histograms")
            print("4. Correlation Heatmap")
            print("5. Variable Dependency")
            print("6. Activity Frequency Distribution Before Balancing")
            print("7. Activity Frequency Distribution After Balancing")
            print("8. Optional Statistics")
            print("9. Models")
            print("10. Quit")
            choice = input("Enter your choice: ")

            if choice == "1":
                df = read_data()
                explore_data(df)
            elif choice == "2":
                df = read_data()
                df = drop_columns(df)
                descriptive_statistics(self.df)
            elif choice == "3":
                df = read_data()
                df = drop_columns(df)
                histograms(df)
            elif choice == "4":
                df = read_data()
                df = drop_columns(df)
                correlation_heatmap(df)
            elif choice == "5":
                df = read_data()
                df = drop_columns(df)
                variable_dependency(df)
            elif choice == "6":
                df = read_data()
                df = drop_columns(df)
                activity_frequency_distribution(df)
            elif choice == "7":
                df = read_data()
                df = drop_columns(df)
                print("Please wait Loading ...")
                x_resampled, y_resampled = oversampling(df)
                print("Please wait Loading ...")
                resampled_activity_freq(y_resampled)
            elif choice == "8":
                self.optional_statistics_menu()
            elif choice == "9":
                self.Menu_Model()
            elif choice == "10":
                break
            else:
                print("Invalid choice. Please try again.")
                
    def Menu_Model(self):
        while True:
            print("\n--- Choose a Model ---")
            print("1. Random Forest Classifier ")
            print("2. Support Vector Machine Classifier")
            print("3. MLP Classifier")
            print("4. Back to Main Menu")
            choice = input("Enter your choice: ")

            if choice == "1":
                df = read_data()
                df = drop_columns(df)
                print("Please wait Loading ...")
                x_resampled, y_resampled = oversampling(df)
                print("Please wait splitting  ...")
                x_train, x_test, y_train, y_test = split_data(x_resampled, y_resampled)
                print("Please wait Scaling ...")
                x_train_scaled, x_test_scaled = scale_data(x_train, x_test)
                print("Please wait predicting ...")
                pickled_rfc_model = load_model('rfc_model.pkl')
                evaluate_classifier(y_test, pickled_rfc_model.predict(x_test_scaled))
            
            elif choice == "2":
                df = read_data()
                df = drop_columns(df)
                print("Please wait Loading ...")
                x_resampled, y_resampled = oversampling(df)
                print("Please wait splitting  ...")
                x_train, x_test, y_train, y_test = split_data(x_resampled, y_resampled)
                print("Please wait Scaling ...")
                x_train_scaled, x_test_scaled = scale_data(x_train, x_test)
                print("Please wait predicting ...")
                pickled_svm_model = load_model('svm_model.pkl')
                evaluate_classifier(y_test, pickled_svm_model.predict(x_test_scaled))
            elif choice == "3":
                df = read_data()
                df = drop_columns(df)
                print("Please wait Loading ...")
                x_resampled, y_resampled = oversampling(df)
                print("Please wait splitting  ...")
                x_train, x_test, y_train, y_test = split_data(x_resampled, y_resampled)
                print("Please wait Scaling ...")
                x_train_scaled, x_test_scaled = scale_data(x_train, x_test)
                print("Please wait predicting ...")
                pickled_mlp_model = load_model('mlp_model.pkl')
                evaluate_classifier(y_test, pickled_mlp_model.predict(x_test_scaled))
            elif choice == "4":
                break
            else:
                print("Invalid choice. Please try again.")

    def optional_statistics_menu(self):
        while True:
            print("\n--- Optional Statistics Menu ---")
            print("1. Compute Variance ")
            print("2. Compute Median")
            print("3. Compute Mean")
            print("4. Compute Standard Deviation")
            print("5. Compute Root Mean Square")
            print("6. Compute Zero Crossings")
            print("7. Compute Sum of Squares")
            print("8. Compute Covariance")
            print("9. Back to Main Menu")
            choice = input("Enter your choice: ")

            if choice == "1":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(variance(df))
            elif choice == "2":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(median(df))
            elif choice == "3":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(mean(df))
            elif choice == "4":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(std(df))
            elif choice == "5":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(rms(df))
            elif choice == "6":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(zero_crossing(df))
            elif choice == "7":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(sum_of_squares(df))
            elif choice == "8":
                df = read_data()
                df = drop_columns(df)
                df = df.drop('activity',axis=1)
                print(covariance(df))
            elif choice == "9":
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    program = TestProgram()
    program.display_menu()
```
